#### Zadanie
Narysuj wykres funkcji $x^3-4x^2+3x-4$ na odcinku $[-1,4]$ niebieską przerywaną linią. Znajdź (wykorzystując poznane dotychczas metody) lokalne maksima i minima tej funkcjina $(-1,4)$ i zaznacz je na rysunku - maksima jako zielone kółka, a minima jako czerwone krzyżyki.

````python
import sympy as sp
x=sp.symbols('x')
f=x**3-4*x**2+3*x-4
D2f=f.diff(x,2)
extrema=sp.solve(f.diff(x))

fig,ax=plt.subplots()
inter=np.linspace(-1,4,100)
g=sp.lambdify(x,f)
ax.plot(inter,g(inter),color='blue',linestyle='--')
for el in extrema:
    if (el>-1 and el<4 and D2f.subs(x,el)>0):
        ax.plot(el,g(el),color='red',marker='x')
    if (el>-1 and el<4 and D2f.subs(x,el)<0):
        ax.plot(el,g(el),color='green',marker='o') 

        
````

### Wykresy 2D danych wielowymiarowych 
* ```plt.quiver``` - strzałki, pole wektorowe $f:R^2\to R^2$
* ```plt.contour``` - wykres konturowy, poziomice funkcji $f:R^2\to R$
* ```plt.contourf``` - z wypełnieniem
* ```plt.imshow``` - wykres jako obrazek



````python
#quiver
x = np.linspace(-2, 2, 10)
y = np.linspace(-2, 2, 10)
X, Y = np.meshgrid(x, y)
fig,ax=plt.subplots()
ax.quiver(X,Y,X*Y,X+Y)
ax.axis([-2,2,-2,2])
````

````python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import sympy as sp
````

````python
fig=plt.figure(figsize=(5,3),facecolor='white')
````

````python
ax=fig.add_axes((0.1,0.1,0.8,0.8))
fig,ax=plt.subplots()
x=sp.symbols('x')
y=x**3+5*x**2+10
z=y.diff(x,1)
w=y.diff(x,2)
extrema=sp.solve(y.diff(x))
t=np.linspace(-5,2,1000)
a=sp.lambdify(x,y)
b=sp.lambdify(x,z)
c=sp.lambdify(x,w)
ax.plot(t,a(t),label='$y(x)=x^3+5x^2+10$',color='green',linewidth=3)
ax.plot(t,b(t),label="y'(x)",color='red')
ax.plot(t,c(t),label='y"(x)',color='blue')
ax.legend(loc='upper left',ncol=3,frameon=False)
plt.axhline(y=0,color='black')
ax.set(xlabel='Oś x', ylabel='Oś y', title='autor: Anna Cabaj')
ax.set_xticks([-4,-2,0,2])
ax.set_yticks([-10,0,10,20,30])
ax.axis([-6,2.5,-12,40]) 
for i in extrema:
  if (i>-4 and i<2 and w.subs(x,i)>0):
    ax.plot(i,a(i),color='g',marker='o')
    d=float(i)
  if (i>-4 and i<2 and w.subs(x,i)<0):
    ax.plot(i,a(i),color='g',marker='o')
    e=float(i)
x=np.linspace(d,e)
ax.fill_between(x,a(x),color='yellow',alpha=0.1)
plt.vlines(x=d,ymin=0,ymax=a(d),color='grey',ls='--',lw=1)
plt.vlines(x=e,ymin=0,ymax=a(e),color='grey',ls='--',lw=1)

plt.show()
````
