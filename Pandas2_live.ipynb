## Apply, groupby, agg

```python
def kategoryzuj_populacje(ludnosc):
  if ludnosc<10000:
    return 'małe miasto'
  elif ludnosc<100000:
    return 'srednie miasto'
  else:
    return 'duze miasto'
```

```python
dane['wielkosc_miasta']=dane.population.apply(kategoryzuj_populacje)
dane.head()
```

```python
#operacje na wierszu
def polozenie(wiersz):
  wiersz['polozenie']=f'({wiersz.lat},{wiersz.lng})'
  return wiersz
dane.apply(polozenie,axis=1)
```

```python
dane.groupby(['country','admin_name']).mean()
```

```python
dane.agg({'population':['min','max'],'lat':['mean']})
```
