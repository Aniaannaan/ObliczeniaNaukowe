Wykresy i wizualizacja danych

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
```

```python
%matplotlib inline 
```

```python
fig=plt.figure(figsize=(5,5),facecolor='lightgreen')
```

```python
ax=fig.add_axes()
ax
```

```python
fig,ax=plt.subplots()
x=np.linspace(-1,4,100)
y=x**3-4*x**2+3*x-4
ax.plot(x,y,label='y', linestyle='--')
```

Pole wrktorrowe

```python
x=np.linspace(-3,3,10)
y=np.linspace(-3,3,10)
X,Y=np.meshgrid(x,y)
fig,ax=plt.subplots()
ax.quiver(X,Y,X**2-Y,1+X+Y**2)
```

Obrazy

```python
chb=np.fromfunction(lambda i, j:i+j+1,(8,8)) %2
fig,ax=plt.subplots()
ax.imshow(chb,cmap='gray')
ax.set_xticks([])
ax.set_yticks([])
ax.plot(np.arange(8),np.arange(8),'x')
```
