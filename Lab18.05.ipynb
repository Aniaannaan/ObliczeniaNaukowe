```python
import skimage
from skimage import data, io, filters
import numpy as np
import matplotlib.pyplot as plt
plt.set_cmap('gray')
```

```python
image=data.camera()
print(image.shape)
image_gauss=filters.gaussian(image,sigma=2)
image_gauss2=filters.gaussian(image,sigma=5)
fig,ax=plt.subplots(1,3,figsize=(20,8))
ax[0].imshow(image)
ax[1].imshow(image_gauss)
ax[2].imshow(image_gauss2)
```

```python
filtr=filters.gaussian(image_gauss2,1)
alpha=6
ostry=image_gauss2+alpha*(image_gauss2.filtr)
fig,size=plr.subplots
```

```python
tekst=data.text()
tekst1=filters.sobel(tekst) #znajdowanie krawdzi
tekst2=filters.sobel_h(tekst) #poziome krawdzie
tekst3=filters.sobel_v(tekst) #pionowe krawdzie
fig,ax=plt.subplots(1,4,figsize=(20,8))
ax[0].imshow(tekst)
ax[1].imshow(tekst1)
ax[2].imshow(tekst2)
ax[3].imshow(tekst3)
```

```python
from skimage import exposure
image=data.camera()
im=exposure.histogram(image)[0]
fig,ax=plt.subplots(1,2,figsize=(20,8))
ax[0].imshow(image)
ax[1].plot(im)
```
